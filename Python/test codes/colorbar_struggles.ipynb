# Using `plt.subplots`

```python
import matplotlib.pyplot as plt
import numpy as np

f, ax = plt.subplots(1,3, width_ratios= [3,1,1], constrained_layout=True)

data0 = np.random.randn(10, 15)
data = np.random.randn(20, 10)
    
a = ax[0].imshow(data0)
b = ax[1].imshow(data)
c = ax[2].imshow(data)

f.colorbar(c, ax=ax[2])
```

```python
from mpl_toolkits import axes_grid1
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

fig = plt.figure()
f, (ax0, ax1, ax2) = plt.subplots(1,3, width_ratios= [3,1,1])#, constrained_layout=True)

data0 = np.random.randn(10, 15)
data = np.random.randn(20, 10)
    
a = ax0.imshow(data0)
b = ax1.imshow(data)
c = ax2.imshow(data)

# https://nbviewer.org/github/mgeier/python-audio/blob/master/plotting/matplotlib-colorbar.ipynb
def add_colorbar(im, aspect=20, pad_fraction=0.5, **kwargs):
    """Add a vertical color bar to an image plot."""
    divider = axes_grid1.make_axes_locatable(im.axes)
    width = axes_grid1.axes_size.AxesY(im.axes, aspect=1./aspect)
    pad = axes_grid1.axes_size.Fraction(pad_fraction, width)
    current_ax = plt.gca()
    cax = divider.append_axes('right', size=width, pad=pad)
    plt.sca(current_ax)
    return im.axes.figure.colorbar(im, cax=cax, **kwargs)

# cbar = fig.colorbar(mappable = c, cax = cax)
add_colorbar(c)
```

# Using `GridSpec` directly

Making an axis for the colorbar

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

fig = plt.figure()
# fig = plt.figure(constrained_layout = True)

gs_outer = gridspec.GridSpec(1, 2, width_ratios= [3, 2], figure = fig)
gs = gs_outer[1].subgridspec(1, 3, width_ratios = [1,1,0.1])

ax0 = plt.subplot(gs_outer[0])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])
cax = plt.subplot(gs[2])

# # alternately:
# ax0 = fig.add_subplot(gs_outer[0])
# ax1 = fig.add_subplot(gs[0])
# ax2 = fig.add_subplot(gs[1])
# cax = fig.add_subplot(gs[2])

data0 = np.random.randn(10, 15)
data = np.random.randn(20, 10)
    
a = ax0.imshow(data0)
b = ax1.imshow(data)
c = ax2.imshow(data)

cbar = fig.colorbar(mappable = c, cax = cax)
```

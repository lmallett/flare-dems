```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import ImageGrid
import matplotlib.colors as colors
import astropy.units as u
import sunpy.visualization.colormaps.color_tables as aiacolormaps

allwaves  = [94,131,171,193,211,335]
indices = [allwaves.index(wave) for wave in allwaves]

colorlist = []
for e in indices:
    colorlist.append(aiacolormaps.aia_color_table(allwaves[e] * u.Angstrom))


# https://stackoverflow.com/questions/41071947/how-to-remove-the-space-between-subplots?rq=1
fig = plt.figure(frameon=False)#figsize=(4, 10))
grid = ImageGrid(fig, 
                 111, # as in plt.subplot(111)
                 nrows_ncols=(2, 3),
                 axes_pad=0,
                 share_all=True,)

for i in range(grid.ngrids):
        ax = grid[i]
        data = np.random.randn(300, 440)
        ax.imshow(data, cmap = colorlist[i])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.axis('off')

plt.margins(0)
# fig.set_size_inches(30, 30)
fig.savefig("test.webp",
            # pad_inches = 0, 
            # bbox_inches= 'tight',
            pil_kwargs = {'lossless':True},
            # dpi=100
)
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt

path = "E:\\emcubes_110809\\emcubes_np\\"
filelist = os.listdir(path)

xregion = [350,651]
yregion = [305,746]

xstart  = xregion[0]
xend    = xregion[1]-1
ystart  = yregion[0]
yend    = yregion[1]-1

fig = plt.figure(frameon=False)#figsize=(4, 10))
grid = ImageGrid(fig, 
                111, # as in plt.subplot(111)
                nrows_ncols=(2, 3),
                axes_pad=0,
                share_all=True,)

for i in range(1):# len(filelist)):
    file = path + filelist[i]
    data = np.load(file)
    data = (data["datacube"])[:, ystart:yend, xstart:xend]

    # fig.set_size_inches(data.shape[1], data.shape[2])
    for i in range(grid.ngrids):
        ax = grid[i]
        ax.imshow(data[i,:,:], interpolation = 'none', origin = 'lower', aspect = 'auto')
        # data = np.random.randn(20, 10)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.axis('off')

        # plt.subplots_adjust(hspace=0, wspace=0)
        plt.tight_layout()
        fig.savefig("test.webp",
                pad_inches = 0, 
                bbox_inches= 'tight',
                pil_kwargs = {'lossless':True},
                # dpi = 1
    )
```

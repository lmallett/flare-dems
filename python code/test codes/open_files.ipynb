```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import ImageGrid
import matplotlib.colors as colors
import astropy.units as u
import sunpy.visualization.colormaps.color_tables as aiacolormaps

allwaves  = [94,131,171,193,211,335]
indices = [allwaves.index(wave) for wave in allwaves]

colorlist = []
for e in indices:
    colorlist.append(aiacolormaps.aia_color_table(allwaves[e] * u.Angstrom))

# https://stackoverflow.com/questions/41071947/how-to-remove-the-space-between-subplots?rq=1
fig = plt.figure(frameon=False)#figsize=(4, 10))
grid = ImageGrid(fig, 
                 111, # as in plt.subplot(111)
                 nrows_ncols=(2, 3),
                 axes_pad=0,
                 share_all=True,)

for i in range(grid.ngrids):
        ax = grid[i]
        data = np.random.randn(300, 440)
        ax.imshow(data, cmap = colorlist[i])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.axis('off')

plt.margins(0)
# fig.set_size_inches(30, 30)
fig.savefig("test.webp",
            # pad_inches = 0, 
            # bbox_inches= 'tight',
            pil_kwargs = {'lossless':True},
            # dpi=100
)
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import sunpy.visualization.colormaps.color_tables as aiacolormaps
import astropy.units as u

colors = [] # list of colormaps
allwaves = [94]

print(aiacolormaps.aia_color_table([allwaves[0]] * u.Angstrom))

for i in range(5):
    colors.append(
        aiacolormaps.aia_color_table([allwaves[i]] * u.Angstrom)
        )


path = "D:\\emcubes_110809\\emcubes_np\\"
filelist = os.listdir(path)

xregion = [350,651]
yregion = [305,746]

xstart  = xregion[0]
xend    = xregion[1]-1
ystart  = yregion[0]
yend    = yregion[1]-1

# Calculate the correct figure size based on data dimensions
data_width = xend - xstart
data_height = yend - ystart
n_cols = 3
n_rows = 2

# Total width = 3 * 10 = 30
# Total height = 2 * 20 = 40
# To maintain aspect ratio: width:height = 30:40 = 3:4
fig_width = 6  # base width
fig_height = fig_width * (n_rows*data_height)/(n_cols*data_width)  # adjust height to maintain 3:4 ratio

fig = plt.figure(figsize=(fig_width, fig_height), dpi=100, constrained_layout=False)
plt.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0, hspace=0)

gs = fig.add_gridspec(2, 3, hspace=-0.1, wspace=-0.1)
axs = gs.subplots(sharex='col', sharey='row').flatten()

for i in range(1):
    file = path + filelist[i]
    data = np.load(file)
    data = (data["datacube"])[:, ystart:yend, xstart:xend]
    print(data.shape)

    # Create mesh grid for pcolormesh
    x = np.arange(data_width + 1)  # One more than data width
    y = np.arange(data_height + 1)  # One more than data height
    X, Y = np.meshgrid(x, y)

    for j in range(6):
        axs[j].pcolormesh(X, Y, data[j], shading='flat', cmap = colors[j])
        axs[j].set_xlim(0, data_width)
        axs[j].set_ylim(0, data_height)
        axs[j].set_xticks([])
        axs[j].set_yticks([])
        axs[j].spines['top'].set_visible(False)
        axs[j].spines['right'].set_visible(False)
        axs[j].spines['bottom'].set_visible(False)
        axs[j].spines['left'].set_visible(False)

fig.savefig("test.webp",
        pad_inches=0,
        bbox_inches=None,
        pil_kwargs={'lossless':True}
)
```

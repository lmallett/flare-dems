```python
import numpy as np
import os
import matplotlib.pyplot as plt
import sunpy.visualization.colormaps.color_tables as aiacolormaps
import astropy.units as u

colors = [] # list of colormaps
allwaves = [94, 131, 171, 193, 211, 335]
for i in range(6):
    colors.append(
        aiacolormaps.aia_color_table(allwaves[i] * u.Angstrom)
        )

path = "D:\\emcubes_110809\\emcubes_np\\"
savdir = "C://Users//Lucien//Documents//School//Research//2023 - DEM Inversion//movies//2011//data_allchannels//"
xregion = [410,601]
yregion = [410,611]

# path = "D:\\emcubes_140910\\emcubes_np\\"
# savdir = "C://Users//Lucien//Documents//School//Research//2023 - DEM Inversion//movies//2014//data_allchannels//"
# xregion = [250,750]
# yregion = [350,600]

filelist = os.listdir(path)

xstart  = xregion[0]
xend    = xregion[1]
ystart  = yregion[0]
yend    = yregion[1]-1

# Calculate the correct figure size based on data dimensions
data_width = xend - xstart
data_height = yend - ystart
height = [data_height, data_height]
width = [data_width, data_width, data_width]

print(data_width, data_height)
print(data_width*3, data_height*2)

px = 1/plt.rcParams['figure.dpi']  # pixel in inches

fig, axs = plt.subplots(2, 3,
        figsize=(sum(width)*px, sum(height)*px),
        gridspec_kw={'height_ratios':height})

titles = ['$94 \mathrm{\AA}$', 
          '$131 \mathrm{\AA}$', 
          '$171 \mathrm{\AA}$', 
          '$193 \mathrm{\AA}$', 
          '$211 \mathrm{\AA}$',
          '$335 \mathrm{\AA}$']

axs = np.reshape(axs, (6))
for i in range(len(filelist)):
    file = filelist[i]
    data = np.load(path + file)
    data = np.clip((data["datacube"])[:, ystart:yend, xstart:xend], a_min=0, a_max=None)**0.25

    print(i)
    for j in range(6):
        axs[j].imshow(data[j],
                      interpolation = 'none',
                      origin = 'lower',
                      cmap = colors[j])
        axs[j].axis('off')
        axs[j].text(0.02, 0.9,
                    titles[j],
                    transform = axs[j].transAxes,
                    c = 'white')

    plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
    plt.show()
    fig.savefig(savdir + file[:-4] + ".png")
```
